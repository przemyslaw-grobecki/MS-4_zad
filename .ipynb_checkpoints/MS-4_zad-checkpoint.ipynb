# Problem A

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as scp
import random
import math
import sys
```

```python
def hipergeometric(m, N, t, M):
    return scp.special.binom(M,m)*scp.special.binom(N-M,t-m)/scp.special.binom(N,t)
```

```python
#M
samples = 1000

N = 50
M = 20
t = 12


labeled = random.sample(range(0, N), M)
unique_picks_empirycal = []
unique_picks_theoretical = []
print(labeled)

for i in range(0,samples):
    temp_ms = random.sample(range(0, N), t)
    #print("real:")
    #print(temp_ms)
    diff = np.setdiff1d(temp_ms, labeled)
    #print("diff:")
    #print(diff)
    unique_picks_empirycal.append(t - diff.size)

#print(unique_picks)
plt.hist(unique_picks_empirycal, density=True, label="empirycal")

x = range(N)
for m in x:
    unique_picks_theoretical.append(hipergeometric(m,N,t,M))

plt.plot(x, unique_picks_theoretical,'r-o', label='theoretical')


plt.legend()
plt.xlabel('Count of desired items in our draw')
plt.ylabel('Probablities')
plt.title('Hypergeometric Distribution Plot')
plt.xlim(0,20)
plt.show()
```

```python
#Init
np.set_printoptions(threshold=sys.maxsize)

PNs = []
labeled = np.array([])
X = range(0,2000)

N = 1000
M = 0
t = 0
m = 0

PN = np.array([1/2000 for _ in X])
PNs.append(PN)  


for i in range(0,19):
    #Go catch some fishes
    t = np.random.randint(40,60)
    fished = np.random.choice(range(0,1000), t, replace=False)
    diff = np.setdiff1d(fished,labeled)
    diff1 = np.setdiff1d(labeled,fished)
    labeled = np.concatenate((labeled, diff), axis=None)
    labeled = np.sort(labeled)
    print("Fished: {}".format(np.sort(fished)))
    print("Difference: {}".format(diff))
    print("All labeled: {}".format(labeled))
    m = t - len(diff)
    M = len(labeled)
    
    print("m: {}".format(m))
    print("t: {}".format(t))
    print("M: {}".format(M))
    
    #Calculate probabilities
    pm = np.array([hipergeometric(m,x,t,M) if x > M+t else PN[x] for x in X])
    pm = np.nan_to_num(pm)
    PN = np.multiply(PN, pm) #Multiply by prior
    PN = np.nan_to_num(PN)
    PN = np.divide(PN, np.sum(PN)) #Normalize
    PN = np.nan_to_num(PN)
    
    
    
    #Add next step
    PNs.append(PN)
    
#plt.ylim(0,1)
#print(PNs)
for i, y in enumerate(PNs):
    plt.plot(X,y, label='{}'.format(i))
plt.legend()
plt.show()
```
